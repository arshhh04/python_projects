```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

data=pd.read_csv("wc.csv")
data.head(3)
# Renaming united states of america to usa
data['1st Team'] = data['1st Team'].replace('United State Of America', 'USA')

data['2nd Team'] = data['2nd Team'].replace('United State Of America', 'USA')
data['Winners'] = data['Winners'].replace('United State Of America', 'USA')
data.head()
```

```python
data.info()
```

```python
new_data = data[data['Top Scorer'] != 'Rain']

high=new_data.sort_values(by='Highest Score', ascending=False)
high.head()
```

```python
req_data=high[:6]
req_data.head()
```

```python
import numpy as np
#converting columns into arrays
plt.figure(figsize=(10,6))

highest_score='Highest Score'
highest_scorer='Top Scorer'
hs_array=np.array(req_data[highest_score])
hsr_array=np.array(req_data[highest_scorer])
sns.lineplot(x=hsr_array,y=hs_array)
plt.title('Highest Score vs top scorer')
plt.xlabel('Top Scorer')
plt.ylabel('Highest Score')
plt.show()
```

```python
new_data = data[data['Best Bowler'] != 'Rain']

high=new_data.sort_values(by='Best Bowler Figure(Wickets Taken)', ascending=False)
high.head()
```

```python
req_data=high[:6]
req_data.head()
```

```python
#converting columns into arrays
plt.figure(figsize=(10,6))
bestbowling='Best Bowler Figure(Wickets Taken)'
bestbowler='Best Bowler'

hs_array=np.array(req_data[bestbowling])
hsr_array=np.array(req_data[bestbowler])
sns.lineplot(x=hsr_array,y=hs_array)
plt.title('Performance of top 5 bowlers')
plt.xlabel('Bowler name')
plt.ylabel('No. of wickets')
plt.show()
```

```python
#barplot for player of the matches
plt.figure(figsize=(10,6))



plt.bar(list(data['Player Of The Match'].value_counts()[0:6].keys()),list(data['Player Of The Match'].value_counts()[0:6]))
plt.xlabel('Name of the player')
plt.ylabel('Number of awards')
plt.title('Top 5 player of the match')
plt.show()
```

```python
#max used venue
data['Venue'].value_counts()



```

```python
#teams with most toss win
toss=data['Toss Winning'].value_counts()
#bargraph representation
plt.figure(figsize=(10,6))
plt.bar(list(toss.keys()[0:6]),list(toss)[0:6])
plt.title('Teams with most toss win')
plt.xlabel('Teams')
plt.ylabel('Number of toss win')
plt.show
```

```python
won_by_runs=data[data['Won by']=='Runs']
won_by_runs.head()
```

```python
#hist repre no. of times a team won when they batted first
#y axis=number of times
#x axis=teams won with runs
plt.figure(figsize=(12, 6))

won_by_runs['Winners'].value_counts()
plt.bar(list(won_by_runs['Winners'].value_counts().keys()),list(won_by_runs['Winners'].value_counts()))
plt.title('No. of times a team won when they batted first')
plt.xlabel('Teams')
plt.ylabel('Number of times')
plt.show()
```

```python
plt.pie(list(won_by_runs['Winners'].value_counts()),labels=list(won_by_runs['Winners'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
